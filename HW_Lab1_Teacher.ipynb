```python
import nltk
```

```python
#4.1
# implementation of inverted index
inverted_index = {"document_count": 0,}

"""filename should be a string indicating the file path"""
def add_to_index(filename):
    document = open(filename, 'r').read()
    words = nltk.word_tokenize(document)
    data_analysis = nltk.FreqDist(words)
    for k, v in data_analysis.items():
        if k in inverted_index:
            inverted_index[k].append((inverted_index["document_count"],v))
            inverted_index[k][0] += 1
        else:
            inverted_index[k] = [1,(inverted_index["document_count"],v)]
    inverted_index["document_count"] += 1
```

```python
add_to_index('filename.txt') 
add_to_index('filename.txt') 
from pprint import pprint
pprint(inverted_index)
```
